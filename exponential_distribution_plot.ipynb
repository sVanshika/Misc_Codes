# PDF for Exponential Distribution:
$f(x) = \lambda e^{-\lambda x}$           (for x > 0)
# PDF for Exponential Distribution for sum of two variables:
z = x1 + x2

$f(z) = \lambda^2 z e^{-\lambda z}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdfExp(x, l):
    return l * np.exp(-l * x)

def pdfExpSum(z, l):
    # z = x1+x2
    return l * l * z * np.exp(-l * z)
```

# Plotting for 1 random variable

```python
x = np.linspace(0, 10, 1000)
pdf = pdfExp(x, 1)

plt.figure(figsize=(16,5))
plt.plot(x, pdf)
plt.title("Exponential distribution of a random variable")
plt.show()
```

# Plotting for sum of n random variable

```python
nums = [2,3,4,5,6,7,8,9,10,23,24,25]
for n in nums:
    z = np.linspace(0, 10, 1000)
    for i in range(n):
        end = np.random.randint(10, 20)
        x = np.linspace(0, end, 1000)
        z += x

    pdf = pdfExpSum(z, 1)
    plt.figure(figsize=(16, 5))
    plt.plot(z, pdf)
    plt.title(f"Exponential distribution for sum of {n} random variables")
    plt.show()
```
